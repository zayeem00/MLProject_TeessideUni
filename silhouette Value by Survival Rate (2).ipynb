```python
from sklearn.cluster import AgglomerativeClustering
```

```python
df_f = pd.DataFrame()
df_m=pd.DataFrame()
range_n_clusters=list(range(2,32))
for n_clusters in range_n_clusters:
    km = AgglomerativeClustering(n_clusters=n_clusters,affinity='euclidean',linkage='ward').fit(data)
    df_m['cluster']=km.labels_
    print(km.labels_)
    df_f = pd.concat([df_m.reset_index(drop=False), df_f], axis=1)
    
    
```
