```python
dt=pd.read_excel('LUSC_250_NoDepth.xlsx')
```

```python
sur=pd.read_excel('survival_LUSC.xlsx')
```

```python
x=dt[["American Joint Committee on Cancer Tumor Stage Code"]]
```

```python
y=sur[['Overall Survival']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
predictions=lm.predict(X_test)
plt.figure(figsize=(20,10))
plt.scatter(y_test,predictions)
plt.xlabel('true value')
plt.ylabel('predicted value')
plt.title("American Joint Committee on Cancer Tumor Stage Code")
```
